```python
import os
folder = "/Users/balthasar/Google Drive/MTL_data/wapo/"
datafiles = [folder + file for file in os.listdir(folder)]
```

```python
from itertools import chain
import json

def get_data(filename):
    with open(filename,'r') as file:
        for line in file:
            yield json.loads(line)

data = chain(*[get_data(f) for f in datafiles])
```

```python
# WAPO Specific

import re 

def get_paragraphs(html):
    soup = BeautifulSoup(html,'html.parser')
    paragraphs = [tag.get_text() for tag in soup.findAll('title')+
                  soup.findAll('meta',property="og:description")+
                  soup.findAll('p')]
    if len(paragraphs) != 0:
        return paragraphs
    else:
        return 'NOCONTENT'

label_re = re.compile('(?<=https://www.washingtonpost.com/)([a-z-]*/){1,}(?=wp)')
def get_label(url):
    label = label_re.search(url)
    if label:
        return label.group().strip('amphtml/')
    else:
        return "NOLABEL"
    
# General   
def unpack(item):
    return tuple(item.values())
```
